```python
from monitor_bot.setting import API_KEY
from slackclient import SlackClient

intro_msg  = json.dumps([{"pretext": "pre-hello",
                          "text": "text-world",
                         "image_url": "https://datadoghq.com/snapshot/path/to/snapshot.png",}])

sc = SlackClient(API_KEY)
sc.api_call(
  "chat.postMessage",
  channel="@suwa",
  text="Hello from Python! :tada:",
  as_user=True,
  attachments=intro_msg
)
```

```python
from io import BytesIO
sc.api_call(
    "files.upload",
    filename="a.png",
    channels="@suwa",
    file=BytesIO(str.encode("https://datadoghq.com/snapshot/path/to/snapshot.png"))
)
```
